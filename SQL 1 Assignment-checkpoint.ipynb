# SQL 1 Assignment

```python
#Read the csv file at the specified location into a DataFrame. Header=None specifies that first row is not header
import pandas as pd
pdTab=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data',header=None,skipinitialspace=True)
```

```python
pdTab
```

```python
#Assign the new column names 
columnList= ['Age','workclass',
'fnlwgt', 'education',
'education-num','marital-status',
'occupation','relationship',
'race','sex','capital-gain',
'capital-loss','hours-per-week','native-country','label']
pdTab.columns = columnList
```

```python
pdTab
```

```python
#Create a table using sqlite3 using the DataFrame created above
import sqlite3
conn=sqlite3.connect("sqladb")
pdTab.to_sql('sqladb',conn)
```

# 1. Select 10 records from the adult sqladb

```python
#1.load the first 10 records
cur=conn.cursor()
sql="Select * from sqladb"
cur.execute(sql)
cur.fetchmany(10)
```

# 2. Show me the average hours per week of all men who are working in private sector

```python
#2. define the sql to fetch average hours per week for sex=Male and workclass=Private from DataFrame and execute to show results
sql='select avg([hours-per-week]) from sqladb where sex="Male" and workclass="Private" group by sex,workclass'
cur.execute(sql)
cur.fetchall()
```

# 3. Show me the frequency table for education, occupation and relationship, separately

```python
#3.a Frequency Table for Education
freqEduc = pd.crosstab(index=pdTab['education'],columns='hours-per-week')
freqEduc
```

```python
#3.b Frequency Table for Occupation
freqOcc = pd.crosstab(index=pdTab['occupation'],columns='hours-per-week')
freqOcc
```

```python
#3.c Frequency Table for Relationship
freqRelat = pd.crosstab(index=pdTab['relationship'],columns='hours-per-week')
freqRelat
```

# 4. Are there any people who are married, working in private sector and having a masters degree

```python
#4. Select count of Age to show the count of people holding Masters degree
sql="select count(Age),workclass,education,[marital-status] from sqladb where workclass='Private'and education='Masters' and [marital-status] like 'Married%' group by education,workclass,[marital-status]"
cur.execute(sql)
cur.fetchall()
```

# 5.What is the average, minimum and maximum age group for people working in different sectors

```python
#5. Average, Minimum and MAximum age for people working in different sectors.
sql='select avg(Age),max(Age),min(Age),workclass from sqladb group by workclass'
cur.execute(sql)
df=pd.DataFrame(cur.fetchall())
df
```

# 6. Calculate age distribution by country

```python
#6. Displaying age distribution by country
pd.crosstab(index=pdTab['native-country'],columns=pdTab['Age'])
```

# 7. Compute a new column as 'Net-Capital-Gain' from the two columns 'capital-gain' and 'capital-loss'

```python
#7. Create a new column using the existing two columns.
sql='select Age,[capital-gain],[capital-loss],([capital-gain]-[capital-loss]) as "Net-Capital-Gain"  from sqladb'
cur.execute(sql)
dfNCG=pd.DataFrame(cur.fetchall(),columns=['Age','Capital Gain','Capital Loss','Net Capital Gain'])
dfNCG
```
